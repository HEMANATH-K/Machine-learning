```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
loan_dataset=pd.read_csv('/content/loan_data.csv')
loan_dataset.head()
```

```python
df=loan_dataset.copy()
df.head()
```

```python
df.isnull().sum()
```

```python
df1=df.dropna()
df1.isnull().sum()
```

```python
l=LabelEncoder()
df1['Loan_Status']=l.fit_transform(df1['Loan_Status'])
df1['Gender']=l.fit_transform(df1['Gender'])
df1['Married']=l.fit_transform(df1['Married'])
df1['Education']=l.fit_transform(df1['Education'])
df1['Property_Area']=l.fit_transform(df1['Property_Area'])
df1['Loan_ID']=l.fit_transform(df1['Loan_ID'])
df1['Dependents']=l.fit_transform(df1['Dependents'])
df1['Self_Employed']=l.fit_transform(df1['Self_Employed'])

df1.head()
```

```python

plt.figure(figsize=(10,10))
sns.heatmap(df1.corr(),annot=True)
```

```python
print(df1.columns)
```

```python
s1 = svm.SVC(kernel='linear')

# Prepare the data
X = df1.drop(columns=['Loan_Status'])
y = df1['Loan_Status']

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train the model
s1.fit(X_train, y_train)

# Make predictions
y_pred = s1.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
